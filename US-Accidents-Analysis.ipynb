### Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time, format='mixed')
```

```python
sns.histplot(df.Start_Time.dt.hour, stat='percent', bins=24)
```

- A high percentage of accidents occur in 6 to 10 am (probably people are in a hurry to get to work)
- next highest percentage is 3 pm to 6 pm 

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

- is the distributuion of accidents by hour the same on weekends as on weekdays. 

```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

- on sundays the peak occurs between 10 am and 3 pm unlike weekdays

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

- much data is missing for 2016. may be even 2017.
